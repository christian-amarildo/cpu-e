# atividade CPU-E: multiplicação e potência

para isso foi utilizado os 7 comandos abaixo:

load: leva o conteúdo do operando para o acc

add: soma o operando com o valor que esta presente no acc
    
sub: subtrai o valor presente no acc pelo operando

store: guarda o valor do acc no endereço indicado no operando

jgt: modifica o valor do ci para o valor do operando 

hlt:encerra o programa
